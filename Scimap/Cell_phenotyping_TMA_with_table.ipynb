```python
import scimap as sm
import anndata as ad
import pandas as pd
```

```python
gates_2_4 = pd.read_csv('/dss/dsshome1/00/ge65tib2/Scimap/gates_2_4.csv')
gates_2_4 = gates_2_4.rename(columns={'2-4__HE__histo': 'gates'})


gates_5_5 = pd.read_csv('/dss/dsshome1/00/ge65tib2/Scimap/gates_5_5.csv')
gates_5_5 = gates_5_5.rename(columns={'5-5__HE__histo': 'gates'})
```

```python
gates_2_4.head()
```

```python
adata_2_4= ad.read_h5ad('/dss/dsshome1/00/ge65tib2/Scimap/2_4_ds_cellpose.h5ad')
adata_2_4.obs['imageid'] = 'img001'

adata_5_5= ad.read_h5ad('/dss/dsshome1/00/ge65tib2/Scimap/5_5_ds_cellpose.h5ad')
adata_5_5.obs['imageid'] = 'img002'
```

```python
adata_2_4 = sm.pp.rescale (adata_2_4, gate=gates_2_4)
```

```python
adata_5_5 = sm.pp.rescale (adata_5_5, gate=gates_5_5)
```

```python
phenotype = pd.read_csv('/dss/dsshome1/00/ge65tib2/Scimap/Phenotyping_table.csv', sep=';')
phenotype.style.format(na_rep='')
```

```python
adata_2_4 = sm.tl.phenotype_cells (adata_2_4, phenotype=phenotype, label="phenotype") 
adata_5_5 = sm.tl.phenotype_cells (adata_5_5, phenotype=phenotype, label="phenotype") 
```
