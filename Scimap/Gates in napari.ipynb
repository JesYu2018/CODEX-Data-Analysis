```python
adata_2_4= ad.read_h5ad('/Users/jessica/Intership_ss2025/scimap/2_4_ds_cellpose.h5ad')
adata_5_5= ad.read_h5ad('/Users/jessica/Intership_ss2025/scimap/5_5_ds_cellpose.h5ad')
adata_2_4.layers["X"] = adata_2_4.X
adata_5_5.layers["X"] = adata_5_5.X

# 查看是否有负值
print("Min value in X:", adata_2_4.X.min())

# 检查是否有 NaN / Inf
print("NaN count:", np.isnan(adata_2_4.X).sum())
print("Inf count:", np.isinf(adata_2_4.X).sum())

# 查看是否有负值
print("Min value in X:", adata_5_5.X.min())

# 检查是否有 NaN / Inf
print("NaN count:", np.isnan(adata_5_5.X).sum())
print("Inf count:", np.isinf(adata_5_5.X).sum())
```

```python
import tifffile as tiff

image_path_2_4 = '/Users/jessica/Intership_ss2025/scimap/2-4__HE__histo.ome.tif'

channels = ['DAPI', 'SOX2', 'FAP', 'cxcr3', 'LRRC15', 'CD20', 'CD68', 'CD45RO', 'CD31', 'PD-1', 'Ki67', 'Keratin 5', 'CD4', 'IFNG', 'SMA', 'CD66', 'Podoplanin', 'CXCR4', 'PGRN', 'Granzyme B', 'Pan-Cytokeratin', 'MARCO', 'CD56', 'HAS1', 'CD107a', 'b-Catenin1', 'HLA-A', 'HLA-DR', 'CD8', 'PDL1', 'TP63', 'GATA6', 'CD34', 'TCF-1', 'Bcl-2', 'PCNA', 'CLDN18', 'CD3e', 'MCT4', 'LAG3']
adata_2_4.obs['imageid'] = '2-4__HE__histo'

# Step 1: 读取图像
img_data = tiff.imread(image_path_2_4)  # 返回形状通常为 (y, x, channels)

# Step 2: 转换为 (channels, y, x)
if img_data.shape[-1] == len(channels):  # 通常为 (y, x, channels)
    img_data = np.transpose(img_data, (2, 0, 1))
    print("Transposed img_data shape:", img_data.shape)

tiff.imwrite("transposed_image.tif", img_data)
sm.pl.napariGater(
    image_path='transposed_image.tif',                  # 或改源码使其使用 img_data 变量
    adata=adata_2_4,
    imageid='imageid',        # 必须与你 adata.obs['imageid'] 匹配
    channel_names=channels,
    x_coordinate='centroid-0',
    y_coordinate='centroid-1',
    layer="X"   # 👈 关键修改               # 你需要在源码中允许传入 img_data 否则需 patch
)
```
